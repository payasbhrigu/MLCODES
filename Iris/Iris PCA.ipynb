```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(iris.drop('species',axis=1))
```

```python
scaled_feat = scaler.transform(iris.drop('species',axis=1))
```

```python
scaled_feat
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_feat)
```

```python
iris_pca = pca.transform(scaled_feat)
```

```python
iris['species'] = iris['species'].map({'setosa':1,'versicolor':2,'virginica':3})
```

```python
iris['species']
```

```python
plt.scatter(iris_pca[:,0],iris_pca[:,1],c=iris['species'],cmap='plasma')
```
