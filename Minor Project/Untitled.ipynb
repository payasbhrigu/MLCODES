```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_excel('MP.xlsx')
data.head()
```

```python
data.reset_index(inplace=True)
data.rename(columns={"index":"Company"},inplace=True)
data.head()
```

```python
#data.drop('level_0',axis=1,inplace=True)
data.head()
```

```python
data['Company'] = data['Company'].apply(lambda x:'.'.join(x.split('.')[1:]))
data['Company'] = data['Company'].apply(lambda x:''.join(x.split('\n')))
data['Company'] = data['Company'].apply(lambda x:''.join(x.split('\t')))
```
